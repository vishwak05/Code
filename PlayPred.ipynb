```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('https://gist.githubusercontent.com/DiogoRibeiro7/c6590d0cf119e87c39e31c21a9c0f3a8/raw/4a8e3da267a0c1f0d650901d8295a5153bde8b21/PlayTennis.csv')
data.head()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in data.select_dtypes(include="object").columns.tolist():
    data[i] = le.fit_transform(data[i])
data.head()
```

```python
data.hist()
plt.show()
```

```python
X=data.drop(['Play Tennis'],axis=1)
y=data['Play Tennis']
```

```python
X.shape,y.shape
```

```python
X.columns
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression(max_iter=150)
model1.fit(X,y)
y1_pred=model1.predict(X)
```

```python
print('Logistic Regression:')
print()
acc1=accuracy_score(y,y1_pred)
print('Accuracy score:',accuracy_score(y,y1_pred))
print()
print('Confusion matrix:\n',confusion_matrix(y,y1_pred))
print()
print('Classification report:\n',classification_report(y,y1_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier(criterion='entropy')
model2.fit(X,y)
y2_pred=model2.predict(X)
```

```python
print('Decision Tree Classifier:')
print()
acc2=accuracy_score(y,y2_pred)
print('Accuracy score:',accuracy_score(y,y2_pred))
print()
print('Confusion matrix:\n',confusion_matrix(y,y2_pred))
print()
print('Classification report:\n',classification_report(y,y2_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model3=KNeighborsClassifier()
model3.fit(X,y)
y3_pred=model3.predict(X)
```

```python
print('K Neighbors Classifier:')
print()
acc3=accuracy_score(y,y3_pred)
print('Accuracy score:',accuracy_score(y,y3_pred))
print()
print('Confusion matrix:\n',confusion_matrix(y,y3_pred))
print()
print('Classification report:\n',classification_report(y,y3_pred))
```

```python
from sklearn.svm import SVC
model4=SVC()
model4.fit(X,y)
y4_pred=model4.predict(X)
```

```python
print('SVM:')
print()
acc4=accuracy_score(y,y4_pred)
print('Accuracy score:',accuracy_score(y,y4_pred))
print()
print('Confusion matrix:\n',confusion_matrix(y,y4_pred))
print()
print('Classification report:\n',classification_report(y,y4_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
model5=GaussianNB()
model5.fit(X,y)
y5_pred=model5.predict(X)
```

```python
print('GaussianNB:')
print()
acc5=accuracy_score(y,y5_pred)
print('Accuracy score:',accuracy_score(y,y5_pred))
print()
print('Confusion matrix:\n',confusion_matrix(y,y5_pred))
print()
print('Classification report:\n',classification_report(y,y5_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
model6=MultinomialNB(alpha=1.0)
model6.fit(X,y)
y6_pred=model6.predict(X)
```

```python
print('MultiNominalNB:')
print()
acc6=accuracy_score(y,y6_pred)
print('Accuracy score:',accuracy_score(y,y6_pred))
print()
print('Confusion matrix:\n',confusion_matrix(y,y6_pred))
print()
print('Classification report:\n',classification_report(y,y6_pred))
```

```python
from sklearn.naive_bayes import BernoulliNB
model7=BernoulliNB(alpha=1.0)
model7.fit(X,y)
y7_pred=model7.predict(X)
```

```python
print('BernoulliNB:')
print()
acc7=accuracy_score(y,y7_pred)
print('Accuracy score:',accuracy_score(y,y7_pred))
print()
print('Confusion matrix:\n',confusion_matrix(y,y7_pred))
print()
print('Classification report:\n',classification_report(y,y7_pred))
```

```python
results=pd.DataFrame({
    'Models':['Logistic Regression','Decision Tree','KNeighbors','SVC','GaussianNB','MultiNomialNB','BernoulliNB'],
    'Score':[acc1*100,acc2*100,acc3*100,acc4*100,acc5*100,acc6*100,acc7*100]
}).sort_values(by='Score',ascending=False)
results
```

```python
results.plot.bar(x='Models',y='Score')
```
